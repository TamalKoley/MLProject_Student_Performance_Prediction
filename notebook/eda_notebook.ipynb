```python
for column in df.columns:
    if df[column].dtype=='O':
        print(f'For column {column} unique values are {df[column].unique()}')
```

```python
numerical_features=[column for column in df.columns if df[column].dtype!='O']
categorical_features=[column for column in df.columns if df[column].dtype=='O']

print(f'We have {len(numerical_features)} numreical features -- {numerical_features}')
print(f'We have {len(categorical_features)} categorical features -- {categorical_features}')
```

```python
df['total_score']=df['reading_score']+df['math_score']+df['writing_score']
df['avarage_score']=df['total_score']/3
```

```python
df.head()
```

```python
math_full_score=df[df['math_score']==100]['gender'].count()
writing_full_score=df[df['writing_score']==100]['gender'].count()
reading_full_score=df[df['reading_score']==100]['gender'].count()

print(f'{math_full_score} students have full marks in math')
print(f'{writing_full_score} students have full marks in writing')
print(f'{reading_full_score} students have full marks in reading')
```

```python
math_low_score=df[df['math_score']<20]['gender'].count()
writing_low_score=df[df['writing_score']<20]['gender'].count()
reading_low_score=df[df['reading_score']<20]['gender'].count()

print(f'{math_low_score} students have low marks in math')
print(f'{writing_low_score} students have low marks in writing')
print(f'{reading_low_score} students have low marks in reading')
```
